# Now I want to split the FirstName and LastName

```python
def split_name(x):
    x["First Name"] = x["Full Name"].apply(lambda x:x.split(" ")[0])
    x["Last Name"] = x["Full Name"].apply(lambda x:x.split(" ")[1])
    return x
```

```python
split_name(load_data())
```
