# iloc

```python
data.iloc[0:10,:]
```

```python
data.loc[0:10,:]
```

# There is a difference between iloc and loc. iloc will take the range of values as n-1.

```python
data.iloc[0:10,0]
```

```python
data.loc[0:10,0]
```

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['x', 'y', 'z'])
```

```python
df
```

```python
print(df.loc['x', 'A'])  # Access a specific element
print(df.loc['x':'y', 'A'])  # Access a range of rows and a specific column
```

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]}, index=['x', 'y', 'z'])
```
