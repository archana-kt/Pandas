# Standard Deviation

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.DataFrame({"x":[10,20,30,40,50]})
```

```python
data
```

```python
type(data)
```

```python
data.info()
```

```python
# To find mean
```

```python
mean = np.sum(data["x"]) / len(data["x"])
```

```python
mean
```

```python
np.mean(data["x"])
```

```python
np.average(data["x"])
```

```python
# To find x-x_bar
```

```python
data["x_bar"] = mean
```

```python
data
```

```python
data["x-x_bar"] = data["x"] - data["x_bar"]
```

```python
data
```

```python
# To find x-x_bar whole square
```

```python
data["square"] = data["x-x_bar"] ** 2
```

```python
data
```

```python
var = np.sum(data["square"] / len(data["x"]))
```

```python
var
```

```python
np.var(data["x"])
```

```python
std = np.sqrt(var)
```

```python
std
```

```python
np.std(data["x"])
```
