```python
data = pd.DataFrame(students)
```

```python
df = pd.DataFrame({'month': [1, 4, 7, 10],
...                    'year': [2012, 2014, 2013, 2014],
...                    'sale': [55, 40, 84, 31]})
```

```python
df
```

```python
df.set_index('month') # index to become the 'month' column
```

```python
df.set_index(['year', 'month']) #  MultiIndex using columns 'year' and 'month'
```

```python
df.set_index([pd.Index([1,2,3,4]), 'year']) #  MultiIndex using an Index and a column
```

```python
import numpy as np
```

```python
df = pd.DataFrame([('bird', 389.0),
...                    ('bird', 24.0),
...                    ('mammal', 80.5),
...                    ('mammal', np.nan)],
...                   index=['falcon', 'parrot', 'lion', 'monkey'],
...                   columns=('class', 'max_speed'))
```

```python
df
```

```python
df.reset_index()
```

```python
df.reset_index(drop=True) # drop` parameter to avoid the old index being added as a column
```

```python
index = pd.MultiIndex.from_tuples([('bird', 'falcon'),
...                                    ('bird', 'parrot'),
...                                    ('mammal', 'lion'),
...                                    ('mammal', 'monkey')],
...                                   names=['class', 'name'])
columns = pd.MultiIndex.from_tuples([('speed', 'max'),
...                                      ('species', 'type')])
```

```python
df = pd.DataFrame([(389.0, 'fly'),
...                    ( 24.0, 'fly'),
...                    ( 80.5, 'run'),
...                    (np.nan, 'jump')],
...                   index=index,
...                   columns=columns)
```

```python
df
```

```python
df.reset_index()
```

```python
# If the index has multiple levels, we can reset a subset of them:

df.reset_index(level='class')
```
