# Pandas dataframe select dtypes

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.DataFrame({"First":np.random.randint(100, size = 20), "Second":np.random.normal(100, 3, 20)})
```

```python
data
```

```python
# 100 - mean
# 3 - standard distribution
# 20 - size
```

```python
data.head()
```

```python
data.info()
```

```python
# First column -> integer
# Second column - >float
```

```python
data["Three"] = "Pass"
```

```python
data
```

```python
data.info()
```

```python
# Now we have 3 datatypes
```

```python
data.info(verbose=False)
```

```python
# If we need only int
```

```python
data.select_dtypes(include='int') # this selects all the integer data types
```

```python
# If we need float
```

```python
data.select_dtypes(include='float')
```

```python
data.select_dtypes(exclude='float')
```

```python
data.info()
```

```python
data.select_dtypes(include=np.number)
```

```python
data.select_dtypes(include='number')
```

```python
data.select_dtypes(exclude='float')
```
