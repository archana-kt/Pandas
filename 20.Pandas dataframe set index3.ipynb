# set_index() with drop attribute

```python
data = pd.DataFrame(students)
```

```python
data
```

```python
data = pd.DataFrame(students, columns = ['Name','Age','City','Country'])
```

```python
data
```

```python
data['Gender'] = ['Female','Male','Male','Female','Female']
```

```python
data
```

```python
data.set_index('Gender',drop=False) # this will create index based on gender and also add gender column
```

```python
data.set_index('Gender',drop=True) # this will create index based on gender and dropped the gender column
```

```python
data.set_index('Gender')
```

# set_index() with append attribute

```python
data
```

```python
data.set_index('Gender',append=True) # this will set index based on Gender with 0,1,2,3,4
```

```python
data.set_index('Gender',append=False) # this will replace the 0,1,2,3,4 with gender
```

```python
data.set_index('Gender',inplace=True) # this will make the existing dataframe replaced and give new dataframe 
```
