```python
data = pd.DataFrame({"Day":pd.to_datetime("27th December 2023") + pd.to_timedelta(np.arange(30),"D"), "Expenses":pd.Series(np.random.randint(50000, size = 30))})
```

```python
data
```

```python
# To know the average expenses of 3 days 
```

```python
data["Expenses"].rolling(window=3).mean()
```

```python
# It gives the average expenditure for the 3 day
```

```python
data["Average_3days_exp"] = data["Expenses"].rolling(window=3).mean()
```

```python
data
```

```python
data["Average_3days_exp"] = data["Expenses"].rolling(window=3).sum()
```

```python
data
```

```python
data["Average_3days_exp"] = data["Expenses"].rolling(window=3).max()
```

```python
data # this gives maximum of three days 
```

```python
data["Average_3days_exp"] = data["Expenses"].rolling(window=3).min()
```

```python
data # this gives minimum of three days
```

```python
# To drop a column
```

```python
data.drop(columns = "Average_3days_exp", axis = 1, inplace = True)
```
